```python
#Standard libraries for data analysis:   
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm, skew
from scipy import stats
import statsmodels.api as sm
import pandas.util.testing as tm
```

```python
example1 = r"/content/sample_data/Narratives.txt"
file1 = open(example1).read().split('"\n"')
```

```python
#Extracting the first 700 complaints narratives 
list1 = file1[:700]
print(list1)
```

```python
!pip install transformers
```

```python
# Summarizing the narratives using summarization pipeline
from transformers import pipeline
summarizer = pipeline("summarization")
summarylist = summarizer(list1, truncation = True, max_length = 110)
summarylist
```

```python
# Allocate a pipeline for sentiment-analysis
sentiment_analyzer = pipeline('sentiment-analysis')
sentimentlist = []
for i in range(len(summarylist)):
  sents = sentiment_analyzer(str(summarylist[i]), truncation = True)
  sentimentlist.append(sents)

sentimentlist
```

```python
# Intent Classification
intent_classifier=pipeline("zero-shot-classification")
labels=['Quit Services','Functionality Concern','Security Concern','Overcharged','Not Benefited']
intentlist = []
for i in range(len(summarylist)):
    intent = intent_classifier(str(summarylist[i]), labels)
    intentlist.append(intent)

intentlist
```

```python
# emotion analysis
emotion_classifier=pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
labels=['happy', 'angry', 'excited','sad', 'fearful','indifferent']
emotionlist = []
for i in range(len(summarylist)):
    emotion = emotion_classifier(str(summarylist[i]), labels)
    emotionlist.append(emotion)

emotionlist
```

```python
summaryData = pd.DataFrame(summarylist)
```

```python
# Dataframe of all the sentiments score for the complaints narratives.
sentimentData = pd.DataFrame()
for i in range (len(sentimentlist)):
  data2 = pd.DataFrame(sentimentlist[i])
  sentimentData = sentimentData.append(data2)

sentimentData = sentimentData.reset_index(drop='True')
sentimentData
```

```python
sentimentData.rename(columns = {'label':'sentiment', 'score':'sentiment score'}, inplace=True)
sentimentData
```

```python
intentData = pd.DataFrame()
for i in range (len(intentlist)):
  data2 = pd.DataFrame(intentlist[i])
  data3 = data2.iloc[:1]
  intentData = intentData.append(data3)

intentData = intentData.reset_index(drop='True')
intentData
```

```python
intentData.rename(columns = {'labels':'intent', 'scores':'intent score'}, inplace=True)
intentData.drop(columns='sequence', inplace=True)
intentData
```

```python
emotionsData.rename(columns = {'labels':'emotions', 'scores':'emotion score'}, inplace=True)
emotionsData.drop(columns='sequence', inplace=True)
emotionsData
```

```python
# Organized Dataframe for summarized text, sentiments and intent score along with the intent label.
datalist = [summaryData,sentimentData, intentData]
final_data = pd.concat(datalist, axis = 1)
final_data
```

```python
print("\n Distribution of sentiments : \n",final_data.sentiment.value_counts())
print("\n Distribution of intent : \n",final_data.intent.value_counts())
#print("\n Distribution of emotions : \n",final_data.emotions.value_counts())
```

```python
final_data.to_csv('TransformerData.csv')
```
